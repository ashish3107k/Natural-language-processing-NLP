```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = ['the glass of milk',
        'the glass of juice',
        'the cup of tea',
        'i am a good boy',
        'i am a good developer',
        'understand']
```

```python
sent
```

```python
len(sent)
```

```python
vocab_size=100
```

```python
vocab_size
```

```python
onehot_repr = [one_hot(word,vocab_size) for word in sent]
```

```python
onehot_repr
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
```

```python
embedding_doc = pad_sequences(onehot_repr, padding = 'pre',maxlen = sent_length)
```

```python
print(embedding_doc)
```

```python
#Features
```

```python
dim = 10
```

```python
model = Sequential()
```

```python
model.add(Embedding (vocab_size,10, input_length = sent_length))
```

```python
model.compile=('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedding_doc)
```
